```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, segmentation, color
from skimage.util import img_as_float
from skimage.segmentation import slic, mark_boundaries
from PIL import Image

# Cargar la imagen
image_path = 'mavic3m/DJI_202411071020_004_Crear-ruta-de-zona2/DJI_20241107102410_0012_D.JPG'  # Cambia esto por el path a tu imagen
image = io.imread(image_path)
image = img_as_float(image)  # Convertir a float

# Aplicar SLIC
n_segments = 20000  # Nmero de superpxeles
compactness = 75   # Parmetro de compactacin
segments = slic(image, n_segments=n_segments, compactness=compactness, start_label=1)

slic_boundaries = mark_boundaries(image, segments)
boundaries = np.array((slic_boundaries * 255.0).astype(np.uint8))

im_boundaries = Image.fromarray(boundaries)
im_boundaries.save("slic_boundaries.jpg")

# Mostrar la imagen segmentada
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagen original
ax[0].imshow(image)
ax[0].set_title("Imagen Original")
ax[0].axis("off")

# Imagen con bordes de superpxeles
ax[1].imshow(boundaries)
ax[1].set_title("Superpxeles con SLIC")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, segmentation, color
from skimage.util import img_as_float
from skimage.transform import resize
from skimage.segmentation import slic, mark_boundaries
from PIL import Image
import cv2
# Cargar la imagen
image_path = 'data/hojas/example1.jpeg'  # Cambia esto por el path a tu imagen
image = io.imread(image_path)

resized_image = resize(image, (image.shape[0] / 24, image.shape[1] / 24), anti_aliasing=True)


resized_image = img_as_float(resized_image)  # Convertir a float

# Aplicar SLIC
n_segments = 10  # Nmero de superpxeles
compactness = 30   # Parmetro de compactacin
segments = slic(resized_image, n_segments=n_segments, compactness=compactness, start_label=1)

slic_boundaries = mark_boundaries(resized_image, segments)
boundaries = np.array((slic_boundaries * 255.0).astype(np.uint8))

im_boundaries = Image.fromarray(boundaries)
im_boundaries.save("leaf_slic_boundaries.jpg")

# Mostrar la imagen segmentada
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagen original
ax[0].imshow(resized_image)
ax[0].set_title("Imagen Original")
ax[0].axis("off")

# Imagen con bordes de superpxeles
ax[1].imshow(boundaries)
ax[1].set_title("Superpxeles con SLIC")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```
